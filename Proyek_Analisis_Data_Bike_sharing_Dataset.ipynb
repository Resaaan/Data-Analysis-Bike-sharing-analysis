## Exploratory Data Analysis (EDA)

##### eksplor tabel hour

```python
hour = pd.read_csv('hour.csv')
hour
```

```python
hourly_usage = hour.groupby(['hr', 'weekday'])[['cnt']].sum().reset_index()
hourly_usage = hourly_usage.sort_values(by='cnt', ascending=False)
hourly_usage
```

Berdasarkan grafik dan tabel tersebut pemakaian sepeda paling banyak terjadi di jam pergi dan pulang kantor untuk weekday dan untuk weekend pemakaian sepeda paling banyak terjadi pada siang hari.

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.pointplot(data=hour, x='hour', y='count', hue='weather', ax=ax)
ax.set(title='Count of bikes during different weathers')
```

Berdasarkan gambar tersebut pemakai sepeda paling tinggi yaitu pada cuaca cerah dan paling rendah pada cuaca hujan lebat.
